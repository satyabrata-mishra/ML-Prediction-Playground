```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
parkinsons_data = pd.read_csv('parkinsons.csv')
parkinsons_data.head()
```

```python
parkinsons_data.shape
```

```python
parkinsons_data.info()
```

```python
parkinsons_data['status'].value_counts()
```

```python
parkinsons_data.groupby('status').mean(numeric_only=True)
```

```python
features = parkinsons_data.drop(columns=['name','status'],axis = 1)
target = parkinsons_data['status']
```

```python
features.head()
```

```python
target.head()
```

```python
features_train,features_test,target_train,target_test = train_test_split(features,target,test_size=0.1,random_state=2)
print(features.shape,features_train.shape,features_test.shape)
print(target.shape,target_train.shape,target_test.shape)
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(features_train)
```

```python
features_train = scaler.transform(features_train)
features_test = scaler.transform(features_test)
```

```python
features_train
```

```python
features_test
```

```python
model = svm.SVC(kernel='linear')
```

```python
model.fit(features_train,target_train)
```

```python
training_pred = model.predict(features_train)
training_acc = accuracy_score(training_pred,target_train)
print('Accuracy on training data is ',training_acc)

testing_pred = model.predict(features_test)
testing_acc = accuracy_score(testing_pred,target_test)
print('Accuracy on training data is ',testing_acc)
```
