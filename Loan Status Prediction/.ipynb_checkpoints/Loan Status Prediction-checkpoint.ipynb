```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import svm
import seaborn as sns
```

```python
loan_dataset = pd.read_csv("loan_data_set.csv")
loan_dataset.head()
```

```python
loan_dataset.shape
```

```python
loan_dataset.describe()
```

```python
loan_dataset.isnull().sum()
```

```python
loan_dataset = loan_dataset.dropna()
loan_dataset.isnull().sum()
```

```python
loan_dataset.head(10)
```

```python
loan_dataset['Gender'] = loan_dataset['Gender'].replace({'Male': 1, 'Female': 0})
loan_dataset['Married'] = loan_dataset['Married'].replace({'Yes': 1, 'No': 0})
loan_dataset['Dependents'] = loan_dataset['Dependents'].replace({'3+': 4})
loan_dataset['Education'] = loan_dataset['Education'].replace({'Graduate': 1, 'Not Graduate': 0})
loan_dataset['Self_Employed'] = loan_dataset['Self_Employed'].replace({'Yes': 1, 'No': 0})
loan_dataset['Property_Area'] = loan_dataset['Property_Area'].replace({'Rural': 0, 'Semiurban':1 ,'Urban': 2})
loan_dataset['Loan_Status'] = loan_dataset['Loan_Status'].replace({'N': 0, 'Y': 1})
loan_dataset.head(10)
```

```python
sns.countplot(x = "Education", hue = 'Loan_Status', data = loan_dataset)
```

```python
sns.countplot(x = "Married", hue = 'Loan_Status', data = loan_dataset)
```

```python
data = loan_dataset.drop(columns=['Loan_ID','Loan_Status'],axis = 1)
label = loan_dataset['Loan_Status']
```

```python
data_train,data_test,label_train,label_test = train_test_split(data,label,test_size = 0.1, stratify = label, random_state = 2)
print(data.shape,data_train.shape,data_test.shape)
print(label.shape,label_train.shape,label_test.shape)
```

```python
classifier = svm.SVC(kernel = 'linear')
classifier.fit(data_train,label_train)
```

```python
training_prediction = classifier.predict(data_train)
training_accuracy = accuracy_score(training_prediction,label_train)
print("Accuracy on training data is ",training_accuracy)

testing_prediction = classifier.predict(data_test)
testing_accuracy = accuracy_score(testing_prediction,label_test)
print("Accuracy on testing data is ",testing_accuracy)
```
