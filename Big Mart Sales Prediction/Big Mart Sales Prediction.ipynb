```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn import metrics
from xgboost import XGBRegressor
```

```python
big_mart = pd.read_csv('Dataset/train.csv')
big_mart.head()
```

```python
big_mart['Item_Weight'] = big_mart['Item_Weight'].fillna(big_mart['Item_Weight'].mean())
```

```python
big_mart['Outlet_Size'].mode()
```

```python
big_mart['Outlet_Size'] = big_mart['Outlet_Size'].fillna(big_mart['Outlet_Size'].mode()[0])
```

```python
big_mart.isnull().sum()
```

```python
big_mart['Item_Identifier'].value_counts()
```

```python
big_mart['Item_Fat_Content'].value_counts()
```

```python
big_mart['Item_Fat_Content'] = big_mart['Item_Fat_Content'].replace({'low fat':'Low Fat','LF':'Low Fat','reg':'Regular'})
```

```python
big_mart['Item_Fat_Content'].value_counts()
```

```python
big_mart['Outlet_Type'].value_counts()
```

```python
encoder = LabelEncoder()
```

```python
big_mart['Item_Identifier'] = encoder.fit_transform(big_mart['Item_Identifier'])
big_mart['Item_Fat_Content'] = encoder.fit_transform(big_mart['Item_Fat_Content'])
big_mart['Item_Type'] = encoder.fit_transform(big_mart['Item_Type'])
big_mart['Outlet_Identifier'] = encoder.fit_transform(big_mart['Outlet_Identifier'])
big_mart['Outlet_Size'] = encoder.fit_transform(big_mart['Outlet_Size'])
big_mart['Outlet_Location_Type'] = encoder.fit_transform(big_mart['Outlet_Location_Type'])
big_mart['Outlet_Type'] = encoder.fit_transform(big_mart['Outlet_Type'])
```

```python
big_mart.head()
```

```python
big_mart.tail()
```

```python
feature = big_mart.drop(columns=['Item_Outlet_Sales'],axis=1)
target = big_mart['Item_Outlet_Sales']
```

```python
feature_train,feature_test,target_train,target_test = train_test_split(feature,target,test_size=0.3,random_state=3)
print(feature.shape,feature_train.shape,feature_test.shape)
print(target.shape,target_train.shape,target_test.shape)
```

```python
regressor = XGBRegressor(n_estimators=300,
    learning_rate=0.05,
    max_depth=5,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=1)
```

```python
regressor.fit(feature_train,target_train)
```

```python
training_prediction = regressor.predict(feature_train)
training_r2 = metrics.r2_score(training_prediction,target_train)
print("R Squared error for training data is ",training_r2)

testing_prediction = regressor.predict(feature_test)
testing_r2 = metrics.r2_score(testing_prediction,target_test)
print("R Squared error for training data is ",testing_r2)
```
